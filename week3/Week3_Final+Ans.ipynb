# Sedimentary rocks

As you may know, there are three fundamental classification of rocks - **Sedimentary, Igneous, and Metamorphic.**
Beyond these super broad classifications, we have much more specific classes based on what is helpful to focus on. In sedimentary rocks, we have clasitc and non-clastic sub classes, which further subdivide further. We look at different classifications based on what type of questions we're trying to answer. Sedimentary rocks are broadly classified into three major categories based on their composition -

## **CLASTIC** (derived from weathering and erosion of land materials)
## **CRYSTALLINE** (form from precipitation of dissolved salts in sea water)
## **BIOCLASTIC** (fragments of living organisms).

# Ternary Diagram

#As a geologist, one often plots qualities of rocks or sediments in a **Ternary Diagram**.
These are pretty useful, because they allow us to plot 3 variables (or dimensions) in a 2 dimensional space. The only thing we need to make sure when plotting these, is the variables need to add to a constant. In most applications for undergrad geology courses, the variables are going to be 3 variables of a rock (like grain size, or chemical composition), and they are going to add to 100.

![alt text](Clastic_Classification.jpg)

Using the above ternary, let's try to figure out how to classify a rock that is 65% sand, 32% gravel, and 3 % mud. This is probably going to be a bit confusing at first, and it's okay if it seems weird the first time through. Try to get to the end of the section, and hopefully everything will start to click towards the end. If we look at the above figure, the top point represents 100% gravel, and the bottom line represents 0 % gravel. The right bottom point represents 100% mud, and the left side represents 0% mud. The bottom left corner represents 100% sand, and the right side represents 0 percent sand. Let's start with gravel. We would start at the very bottom line (this represents 0% gravel, and go straight up the triangle, until we reach 32% the height of the triangle, and we draw a straight line parellel with the bottom (the blue line). Next, lets do the mud axis, and so we know it's 3% mud. So we go to very left side, because that represents 0% mud, and we go straight to the bottom right corner, 3% of the distance to the corner and draw a line that is parallel to the left side. (this is the green line). 

The point where these two lines intersect is the composition of the rock. If we draw a line that is perpendicular to the right side (red line), the point is intersected, and the sand value is given at about 65% on the bottom axis. 

We can see that this rock would be a Sandy Conglomerate.

# Normalizing the parameters:

Often times, there will be instances where the components you are trying to measure do not add to 100 (or whatever constant you need it to add up to). In the example above,  the sand, gravel, and mud percentagess added up to 100. What if instead, the sample we were looking at was 15% gravel, 20% sand, 30% mud, and 35% crystalline? And we still wanted to plot on the mud/sand/gravel ternary diagram? It wouldnt work until we normalize the data. What we need to do is to only look at the gravel, sand, and mud paramteres and get rid of the crystalline.

So how do we normalize data? What we would do is say, gravel, sand, and mud make up 100% of what we want to look at, so 65%(15+20+30) of the rock is what we want to look at. 15/65 = 23%, 20/65 = 31%, and 30/65 =  46%. Now look, 23 + 31 + 46 adds up to 100. The data is now normalized. The normalization equation is as follows:

\begin{align}
\textrm{Normalized Total (in ternary plots, it is the sum of the 3 paramters you want to plot)} = X_1 + X_2 + X_3 = N_t
\end{align}


\begin{align}
\textrm{New Normalized Parameters}= \begin{cases}
X_1/N_t = X_{N1} \\
X_2/N_t = X_{N2} \\
X_3/N_t = X_{N3} \\
\end{cases}
\end{align}

\begin{align}
X_{N1} + X_{N2} + X_{N3} = 100
\end{align}

# One more type of ternary (some say it's almost a quad(t)ernary lol)

Up till now, we've sorta just focused on one type of ternary plot - the clast size of clastic sediments/rocks. In addition to looking at clast size, we'll be looking at compoisition. In this next ternary, we're only going to be looking at feldspar content, quartz content, and lithic content in the main body of the ternary (this makes up a **QFL ternary** and we're going to add one more dimension - percent matrix. It ends up giving the ternary an elongate-triangle shape.

Here is a chart of another ternary we're gonna be looking at - it's called the **Pettijohn classification** and only really applies to sandstones.The QFL ternary only applies to sandstones, and for this exersize, we will consider rocks that have greater than 50% sand. Similar to the Gravel, Sand, Mud corners, the QFL corners correspond to Quartz, Feldspar, and Lithics. You can also ignore that tiny ternary in the bottom right corner of this figure (it applies to uncommon mafic sediments like the ones that form from black sand beaches), we're only interested in the toblerone-looking plot.
![alt text](Sandstone_classification.jpg)

So let's say we had a rock, and it had 66.66% sand, and 33.33% mud, with composition of 98% quartz, 1% feldspar, and 1% lithics. So it would plot at the bottom line (because of 0% gravel), 2/3's of the way to the sand corner. Go ahead and print out a blank ternary and try to map it out yourself. Since the sand and mud composition adds to 100 (or 99.99%, lets just say 100%), our compositions are already normalized. Since the normalized sand composition is greater than 50%, we can use the Pettijohn classification. It would be reallllly close to the quart corner, since it's 98% quartz. Then we look at the **matrix %** (which is just the normalized mud composition), and we see it is 33%, so it lies betweeen 15% and 75%. We would then classify this rock as Quartz Wacke.

# Question 4 : Using the constraints we've defined in the lesson, can we ever plot a Mudrock on the Pettijohn ternary? Why or why not?

No, because we defined a "sandstone" that can be used in the Pettijohn classification to have a normalized sand value of 50%. That means the normalized mud value can never be more than 50%. The normalized mud value is the same as the matrix percentage, so the matrix percentage can be a maximum of 50%. Mudrocks need a matrix percentage of 75%. Matrix is usually defined as the pore space between sand grains, and typically consist of really fine grains (mud and silt-sized particles) as well as precipitate minerals. We didnt include precipitates in our calculation, so 50% matrix si the highest we can go given the constraints for this exercise.
